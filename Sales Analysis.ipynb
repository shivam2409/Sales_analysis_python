```python
df.replace({'M': 'Male', 'F': 'Female'}, inplace=True)
```

```python
salesbyGender = sns.countplot(x='Gender',data=df)

for bars in salesbyGender.containers:
    salesbyGender.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
amountByGender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=amountByGender)
```

```python
amountByGender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=amountByGender)

for bars in amountByGender.containers:
    amountByGender.bar_label(bars)
```

```python
ordersByAgeGroup = sns.countplot(data =df,x= 'Age Group',hue='Gender')
```

```python
ordersByAgeGroup = sns.countplot(data =df,x= 'Age Group',hue='Gender')
for bars in ordersByAgeGroup.containers:
    ordersByAgeGroup.bar_label(bars)
```

```python
ordersByStates = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(data = ordersByStates,x ='State',y = 'Orders')
```

```python
amountByStates = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(data = amountByStates,x ='State',y = 'Amount')
```

```python
salesByOccupation = sns.countplot(data=df,x='Occupation')

for bars in salesByOccupation.containers:
    salesByOccupation.bar_label(bars)
```

```python
amountByOccupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.barplot(data = amountByOccupation,x= 'Occupation',y = 'Amount')

```
